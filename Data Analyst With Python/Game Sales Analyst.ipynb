```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
game = pd.read_csv('vgsales.csv').dropna(how='all')
game.head(10)
```

```python
game = game.dropna(subset=['Year'])
```

```python
game['Publisher'].unique()
```

```python
game['Publisher'] = game['Publisher'].fillna('Unknown')
```

```python
game.head(10)
```

```python
game.set_index('Rank', inplace=True)
```

```python
game.head(10)
```

```python
game['Year'] = game['Year'].astype(int)
```

```python
game.columns
```

```python
top_5_genre_games_most_sales  = game['Genre'].value_counts().head(5).index
game_top_5 = game[game['Genre'].isin(top_5_genre_games_most_sales)]
plt.figure(figsize=(10,10))
sns.countplot(x='Genre',data=game_top_5,palette='pastel')
plt.title('Top 5 Genre Games Most Sales')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=40)
plt.show()
```

```python
game.columns
```

```python
sales_total = {
    'NA': game['NA_Sales'].sum(),
    'EU': game['EU_Sales'].sum(),
    'JP': game['JP_Sales'].sum(),
    'Other': game ['Other_Sales'].sum()
}

plt.figure(figsize=(10,10))
plt.pie(sales_total.values(),labels=sales_total.keys(),autopct='%1.1f%%',startangle=140)
plt.title('Distribution Sales Regional or Source')
plt.ylabel('')
plt.show()
```

```python
best_selling_by_year_release = game['Year'].value_counts().sort_index()
plt.plot(best_selling_by_year_release.index,best_selling_by_year_release.values,marker='o',ls='--')
plt.title('Best Selling By Year Release')
plt.ylabel('Count')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

```python
game['Publisher'].unique()
```

```python
top_10_publisher_best_sales = game.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,10))
top_10_publisher_best_sales.plot(kind='barh',color='blue')
plt.title('Top 10 Publisher The Best Sell')
plt.xlabel('Publisher')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
