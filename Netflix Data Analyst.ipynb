###### NETFLIX EDA DATA ANALYST

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv('netflix_titles.csv').dropna(how='all')
netflix.head(10)
```

```python
netflix.shape
```

```python
netflix.info()
```

```python
netflix.isnull().sum().sort_values(ascending=False)
```

```python
netflix.duplicated().sum()
```

```python
df = netflix.copy()
```

```python
df.head(10)
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],format = 'mixed')
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['year_added'].astype(int)
df['month_added'] = df['month_added'].astype(int)
```

```python
df.head(10)
```

```python
df = df.drop('date_added',axis=1)
```

```python
df.head(10)
```

```python
type_distribution = df['type'].value_counts()
plt.figure(figsize=(8,5))
sns.countplot(x='type',data=df,palette='pastel')
plt.title('Type Distribution')
plt.ylabel('Count Of Distribution')
plt.xlabel('Type')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
flow_production_movie_a_year = df['year_added'].value_counts().sort_index()
plt.plot(flow_production_movie_a_year, ls='--', marker='o')
plt.title('Production Movie a Year')
plt.ylabel('Counts a Movie')
plt.xlabel('Year')
plt.show()
```

```python
top_5_distribution_rating = df['rating'].value_counts().head(5)
plt.figure(figsize=(8,5))
plt.pie(top_5_distribution_rating,autopct='%1.1f%%',startangle=140,labels=top_5_distribution_rating.index)
plt.title('Distribution Ration')
plt.ylabel('')
plt.show()
```

```python
distribution_rating = df['rating'].value_counts()
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=df,palette='pastel')
plt.title('Relation Between rating and type')
plt.ylabel('Count')
plt.xlabel('rating')
plt.show()
```
